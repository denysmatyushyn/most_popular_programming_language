```python
most_popular_all_time = df.iloc[-1, 1:-1].sort_values(ascending=False)
plt.figure(figsize = (10,7))
sns.barplot(x = most_popular_all_time.index,
            y = most_popular_all_time.values)
plt.xticks(rotation = 90)
plt.title('Most popular programming languages now (2022 year)')
plt.show()
```

```python
df['Year'] =  pd.DatetimeIndex(df['Date']).year
median_by_year = df.groupby(['Year'])[most_popular_all_time.index[:10]].median()
median_by_year.head(len(df))
```

```python
median_by_year[most_popular_all_time.index[:10]].plot(figsize=(12, 9))
plt.title('The dynamics of the popularity of TOP-10 programming languages', fontsize=16)
plt.ylabel('Сoefficient', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.grid(which = 'major', color = 'k', linestyle = '-.', linewidth=0.5)
plt.show()
```
